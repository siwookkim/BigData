```python
import pandas as pd
```

```python
crime_seoul = pd.read_excel('../data/Report (1).xls', encoding = 'UTF-8')
```

```python
crime_seoul.head()
```

```python
del crime_seoul['기타']
```

```python
crime_seoul.sort_values(by='기간', ascending=False)
```

```python
crime_seoul.rename(columns={crime_seoul.columns[2]: '범죄건수'}, inplace=True)
```

```python
crime_seoul.head()
```

```python
crime_seoul.drop([0], inplace = True)
```

```python
crime_seoul.head()
```

```python
crime_seoul['노상범죄율'] = crime_seoul['노상']/crime_seoul['범죄건수']*100
```

```python
crime_seoul.head()
```

```python
del crime_seoul['기간']
```

```python
crime_seoul.head()
```

```python
crime_seoul['아파트연립범죄율'] = crime_seoul['아파트 연립다세대']/crime_seoul['범죄건수']*100
crime_seoul['단독주택범죄율'] = crime_seoul['단독주택']/crime_seoul['범죄건수']*100
crime_seoul['유흥접객범죄율'] = crime_seoul['유흥접객업소']/crime_seoul['범죄건수']*100
```

```python
crime_seoul.head()
crime_seoul.set_index('구분', inplace = True)
```
